```python
import ipywidgets as widgets
from IPython.display import display

class ToDoListGUI:
    def __init__(self):
        self.tasks = []

        self.task_input = widgets.Text(placeholder='Enter a task')
        self.add_button = widgets.Button(description='Add Task')
        self.task_list = widgets.SelectMultiple(options=self.tasks, layout={'height': '200px'})
        self.delete_button = widgets.Button(description='Delete Task')
        self.complete_button = widgets.Button(description='Mark as Completed')

        self.add_button.on_click(self.add_task)
        self.delete_button.on_click(self.delete_task)
        self.complete_button.on_click(self.mark_completed)

        display(widgets.VBox([widgets.Label('To-Do List'),
                              self.task_input,
                              self.add_button,
                              self.task_list,
                              widgets.HBox([self.delete_button, self.complete_button])]))

    def add_task(self, btn):
        task = self.task_input.value
        if task:
            self.tasks.append(task)
            self.task_list.options = self.tasks
            self.task_input.value = ''

    def delete_task(self, btn):
        selected_tasks = self.task_list.value
        if selected_tasks:
            for task in selected_tasks:
                self.tasks.remove(task)
            self.task_list.options = self.tasks

    def mark_completed(self, btn):
        selected_tasks = self.task_list.value
        if selected_tasks:
            for task in selected_tasks:
                index = self.tasks.index(task)
                self.tasks[index] = f"{task} (Completed)"
            self.task_list.options = self.tasks

todo_gui = ToDoListGUI()
```
